```python
import pandas as pd

# read in the data
df = pd.read_csv('gta_i_0501_0710.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
# separate the TradingDate column to Year, Month, Day which is timestamp
df['TradingDate'].head()
# transform the data type of TradingDate to string
df['TradingDate'] = df['TradingDate'].astype(str)
# split the TradingDate column to Year, Month, Day
df['Year'] = df['TradingDate'].str.slice(0,4)
df['Month'] = df['TradingDate'].str.slice(4,6)
df['Day'] = df['TradingDate'].str.slice(6,8)
# transform the data type of Year, Month, Day to int
df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)
df.drop('TradingDate', axis=1, inplace=True)
df.head()
```

```python
new_df = df[['AskPrice1', 'BidPrice1', 'AskVolume1', 'BidVolume1', 'AskPrice5',
       'AskPrice4', 'AskPrice3', 'AskPrice2', 'BidPrice2', 'BidPrice3',
       'BidPrice4', 'BidPrice5', 'AskVolume5', 'AskVolume4', 'AskVolume3',
       'AskVolume2', 'BidVolume2', 'BidVolume3', 'BidVolume4', 'BidVolume5', 'Year', 'Month', 'Day']]

new_df['date'] = pd.to_datetime(new_df[['Year', 'Month', 'Day']])
new_df['trading_day'] = new_df.groupby('date').ngroup() + 1
new_df.drop(['date','Year', 'Month', 'Day'], axis=1, inplace=True)
new_df.head()
```

```python
new_df.to_csv('cleaned_dataset.csv', index=False)
```
